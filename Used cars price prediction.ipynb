```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.pandas.set_option('display.max_columns',None)  #to display all the columns
```

```python
cars=pd.read_csv('used_cars.csv')
cars
```

```python
# Remove 'mi' '$'
cars['milage'] = cars['milage'].str.replace('mi.', '')
cars['milage'] = cars['milage'].str.replace(',', '')
cars['price'] = cars['price'].str.replace('$', '')
cars['price'] = cars['price'].str.replace(',', '')
cars
```

```python
# Convert to numerical datatype
cars['milage'] = pd.to_numeric(cars['milage'])
cars['price'] = pd.to_numeric(cars['price'])
```

```python
#numerical catagories
num_val=[val for val in cars.columns if cars[val].dtypes!='O']
print(num_val)
```

```python
null_val=[val for val in cars.columns if cars[val].isnull().sum()>0.2]
for val in null_val:
    print('{} column has {} null values'.format(val,round(cars[val].isnull().mean(),4)))
```

```python
num_null_val=[val for val in num_val if cars[val].isnull().sum()>0.2]
for val in num_null_val:
    print('{} column has {} null values'.format(val,round(cars[val].isnull().mean(),4)))
```

```python
cat_val=[val for val in cars.columns if cars[val].dtypes=='O']
print(cat_val)
```

```python
cat_null_val=[val for val in cat_val if cars[val].isnull().sum()>0.0001]
for val in cat_null_val:
    print('{} column has {} null values'.format(val,round(cars[val].isnull().mean(),4)))
```

```python
for val in null_val:
    cars.groupby(val)['price'].median().plot.bar()
    plt.xlabel(val)
    plt.ylabel('price')
    plt.show()
```

```python
for val in num_val:
    cars.groupby(val)['price'].median().plot.bar()
    plt.xlabel(val)
    plt.ylabel('price')
    plt.show()
```

```python
for val in cat_val:
    cars.groupby(val)['price'].median().plot.bar()
    plt.xlabel(val)
    plt.ylabel('price')
    plt.show()
```

```python
#identifing the temporary values
temp_val=[val for val in cars.columns if 'year' in val]
print(temp_val)
```

```python
for val in temp_val:
    print(val,cars[val].unique())
```

```python
for val in temp_val:
    cars.groupby(val)['price'].median().plot()
    plt.xlabel(val)
    plt.ylabel('price')
    plt.show()
```

```python
#discrete values
dist_val=[val for val in num_val if len(cars[val].unique())<25 and val not in temp_val]
print(dist_val)
```

```python
for val in dist_val:
    cars.groupby(val)['price'].median().plot.bar()
    plt.xlabel(val)
    plt.ylabel('price')
    plt.show()
```

```python
#contineous variables
contin_val=[val for val in num_val if val not in dist_val and val not in temp_val]
print(contin_val)
```

```python
for val in contin_val:
    cars[val].hist(bins=25)
    plt.xlabel(val)
    plt.ylabel("Count")
    plt.title(val)
    plt.show()
```

```python
#finding the outliers
for val in contin_val:
    dataset=cars.copy()
    if 0 in dataset[val].unique():
        pass
    else:
        dataset[val]=np.log(dataset[val])
        dataset.boxplot(column=val)
        plt.ylabel(val)
        plt.title(val)
        plt.show()
```

```python
#replacing the null numerical values
for val in num_null_val:
    dataset=cars.copy()
    median_val=dataset[val].median()
    if 0 in dataset[val].unique():
        pass
    else:
        dataset[val].fillna(median_val,inplace=True)
```

```python
dataset.isnull().sum()>1
```

```python
#handling the missing cat values
for val in cat_null_val:
    cars[val].fillna('Missing',inplace=True)
cars.isnull().sum()>0.0001
```

```python
#converting the values into gaussian values for training and testing
for val in num_val:
    cars[val]=np.log(cars[val])
cars
```

```python
#finding the rare catogorical values
for val in cat_val:
    if val not in ['model']:
        temp=cars.groupby(val)['price'].count()/len(cars)
        temp_df=temp[temp>0.01].index
        cars[val]=np.where(cars[val].isin(temp_df),cars[val],'Rare_var')
```

```python
cars.head(50)
```

```python
#replacing catagorical values with labels
for val in cat_val:
    labels_ordered=cars.groupby([val])['price'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    cars[val]=cars[val].map(labels_ordered)
```

```python
cars.head(50)
```

```python
#feature engineering
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
train_dataset=pd.read_csv('x_train_car.csv')
```

```python
train_dataset.head()
```

```python
## Capture the dependent feature
y_train=train_dataset[['price']]
## drop dependent feature from dataset
X_train=train_dataset.drop(['price'],axis=1)
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_train.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
selected_feat
```

```python
X_train=X_train[selected_feat]
```
